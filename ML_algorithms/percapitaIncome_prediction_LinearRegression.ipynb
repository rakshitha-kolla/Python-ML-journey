```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

#### For fitting the modle x always should be 2d and y can be 1d

```python
X=df.iloc[:,[0]]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_train
```

```python
plt.scatter(X_test,y_test,color="red",marker="*")
plt.xlabel("YEAR")
plt.ylabel("Per capita income")
plt.show()
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
plt.scatter(X_test,y_test,color="red",marker="*")
plt.xlabel("YEAR")
plt.ylabel("Per capita income")
plt.plot(X_test,y_pred,color="blue")
plt.show()
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
print(cross_val_score(LinearRegression(),X,y))
```

```python
kf=KFold(n_splits=5,shuffle=True, random_state=42)
```

```python
print(cross_val_score(LinearRegression(),X,y,cv=kf,scoring="r2"))
```

```python
print(np.average(cross_val_score(LinearRegression(),X,y,cv=kf)))
```
