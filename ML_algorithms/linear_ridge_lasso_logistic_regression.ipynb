# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
print("mse=",mean_squared_error(y_test,y_pred))
print("mae=",mean_absolute_error(y_test,y_pred))
print("r2_score=",r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lin_reg=LinearRegression()
```

```python
mse=cross_val_score(lin_reg,X,y,scoring="neg_mean_squared_error",cv=5)
```

```python
mse
```

```python
mean_mse=np.mean(mse)
mean_mse
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=100)
```

```python
ridge.fit(x_train,y_train)
```

```python
y_pred=ridge.predict(x_test)
```

```python
print("r2_score=",r2_score(y_test,y_pred))
```

```python
# Tuning Alpha with GridSearchCV
# you can use GridSearchCv to find optimal value of alpha
```

```python
from sklearn.model_selection import GridSearchCV
param={"alpha":[1e-15,1e-13,1e-10,1e-8,1e-5,1e-2,1e-1,1,10,20,100]}
ridge=Ridge()
ridge_regressor=GridSearchCV(ridge,param,cv=5,scoring="neg_mean_squared_error")
ridge_regressor.fit(x_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Logistic Regression

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
```

```python
dataset=load_breast_cancer()
```

```python
dataset
```

```python
df=pd.DataFrame(dataset["data"],columns=dataset["feature_names"])
```

```python
df
```

```python
df["target"]=dataset["target"]
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred=log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# using GridSearchCV
logistic=LogisticRegression(max_iter=100,C=100)
param_grid={"C":[1,5,10,100],"max_iter":[100,150]}
model=GridSearchCV(logistic,param_grid=param_grid,scoring="f1",cv=5)
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
