```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
np.random.rand(200,1)
```

np.random.rand(200,1)

```python
np.random.rand(200,1).shape
```

```python
X=6*np.random.rand(200,1)-3
y=0.8 * X**2 + 0.9 * X + 2 + np.random.rand(200,1)
#y=0.8x^2+0.9X+2
```

```python
X
```

```python
y
```

```python
plt.plot(X,y,"b",marker=".",linestyle="")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
```

```python
plt.plot(X_train,lr.predict(X_train),color="r")
plt.plot(X_train,y_train,"b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# we clearly see line is linear but there eis curve
# so we need to introduce polynomial features so we get curve that joins points
```

```python
# Applying polynomial Linear Regression
# degree 2
poly=PolynomialFeatures(degree=2)
X_train_trans=poly.fit_transform(X_train)
X_test_trans=poly.transform(X_test)
```

```python
print(X_train[0])
print(X_train_trans[0])
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train_trans,y_train)
```

```python
y_pred=lr.predict(X_test_trans)
```

```python
print(y_pred.shape)
```

```python
print(r2_score(y_test,y_pred))
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
y_new=lr.predict(X_new_poly)
```

```python
plt.plot(X_train,y_train,"b.",label="training points")
plt.plot(X_test,y_test,"g.",label="testing points")
plt.plot(X_new,y_new,"r-",linewidth=2,label="predictions")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
