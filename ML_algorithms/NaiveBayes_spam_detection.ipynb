```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
data=pd.read_csv("spam.csv")
data
```

```python
data.isnull().any()
```

```python
dummy=pd.get_dummies(data.Category)
```

```python
column=dummy.columns
column
```

```python
data[column]=dummy
```

```python
data
```

```python
data.drop(["Category","ham"],axis="columns",inplace=True)
```

```python
data
```

```python
data.isnull().any()
```

# as ml model does not take character we need to transform message also

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X=v.fit_transform(data["Message"])
```

```python
X=X.toarray()
```

```python
y=data["spam"]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model=MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
emails=[
    "Hey mohan, can we get together to watch football game tomarrow?",
    "Upto 20% discount on parking, exclusive offer just for ypu. Dont miss this reward!"
]
```

```python
em=v.transform(emails).toarray()
em
```

```python
model.predict(em)
```

## Using Pipelining

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([
        ("vectorizer",CountVectorizer()),
        ("nb",MultinomialNB())
])
```

```python
clf.fit(data["Message"],data["spam"])
```

```python
clf.predict(emails)
```
