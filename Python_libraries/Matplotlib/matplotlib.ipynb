## 2D plots

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]
chor_bazar=[100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id,online_price,color="navy",marker="*",linestyle="dashdot",markersize=15,linewidth=3)
plt.xlabel("Products")
plt.ylabel("Prices")
plt.title("Product vs Proce ")
plt.show()
```

```python
## if we need to plot for more than 1 lien
```

```python
plt.plot(product_id,online_price,color="navy",marker="*",linestyle="dashdot",markersize=15,linewidth=3,label="online prices")
plt.plot(product_id,offline_price,color="red",marker="*",linestyle="dashdot",markersize=15,linewidth=3,label="ofline prices")
plt.plot(product_id,chor_bazar,color="yellow",marker="*",linestyle="dashdot",markersize=15,linewidth=3,label="chor bazar proces")
plt.xlabel("Products")
plt.ylabel("Prices")
plt.title("Product vs Proce ")
plt.legend()
plt.show()
```

## Xticks,yticks


```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
x2=np.arange(0,4.5,0.5)
## resize dpi is pixels for inch
plt.figure(figsize=(5,3),dpi=200)
plt.plot(x,y,marker="*",label="3",color="blue",linewidth=3,linestyle="dashdot")
plt.plot(x2[:6],x2[:6]**2,label="x^2",color="red")
plt.plot(x2[5:],x2[5:]**2,linestyle="dashed",color="red")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Our first graph",fontdict={"fontsize": 30})
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show() 
```

## Bar Chart

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch("*")
bars[2].set_hatch("o")
plt.figure(figsize=(6,4))
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)
plt.figure(figsize=(6,4))
patterns=['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop())
plt.show()
```

## Pie Chart2

```python
fifa.Weight
```

```python
fifa.Weight=[int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight >=125) & (fifa.Weight <150)].count()[0]
medium=fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy=fifa.loc[(fifa.Weight >= 200)].count()[0]
```

```python
## explode is used to split pie chart
```

```python
weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['Under 125',"125-150","150-175","175-200","Over 200"]
plt.style.use("ggplot")
explode=[.4,.2,0,0,.4]
plt.pie(weights,labels=labels,autopct="%.2f%%",explode=explode)
plt.show()
```
