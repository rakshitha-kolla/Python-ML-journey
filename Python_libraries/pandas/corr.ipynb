## Corr() Function

```python
## it used to find the pairwise correaltion of all columns in dataframe 
## NaN value are automatically excluded
## how strongly the variables are related
## corr works on only numerical data
```

```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv("matches.csv")
match.head()
```

```python
match.corr(numeric_only=True)
```
