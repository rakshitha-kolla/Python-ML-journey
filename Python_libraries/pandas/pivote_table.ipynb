## Pivote table

```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv("matches.csv")
delivery=pd.read_csv("deliveries.csv")
```

```python
match.head()
```

```python
delivery.head()
```

```python
mask=delivery["batsman_runs"]==6
six=delivery[mask]
six.head()
```

```python
six.shape
```

```python
pt=six.pivot_table(index="over",columns="batting_team",values="batsman_runs",aggfunc="count")
pt
```

```python
## convert pivote table into heat map
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
