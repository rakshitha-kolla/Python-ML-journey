# Merge

#### ->we need to fing orange cap holders of every year
#### ->orange cap holder is a person who scores highest run in season

```python
import pandas as pd
import numpy as np
```

```python
delivery=pd.read_csv("deliveries.csv")
delivery
```

```python
## we observe there no column that represents season or year
## now we need to merge datasets to get relevent information
## we observe in deliveries.csv and matches.csv id is unique column so we can merge using that column
```

```python
matches=pd.read_csv("matches.csv")
matches
```

```python
## marge syntax
## table.merge(name of second table,left_on="1st table column name",right_on="2nd table column name")
## by default how='inner' (inner considers those which are present in both) ,we can change to outer or left or right join
```

```python
merged=delivery.merge(matches,left_on="match_id",right_on="id")
```

```python
merged
```

```python
merged.shape
```

```python
delivery.shape
```

```python
matches.shape
```

```python
new=merged.groupby(["batsman","season"])["batsman_runs"].sum().copy()
```

```python
new
```

```python
new=new["season"].sort_values(ascending=False)
```

```python
new.index
```

```python
new=new.reset_index()
```

```python
new
```

```python
new.drop_duplicates(subset="season",keep="first",inplace=True)
```
