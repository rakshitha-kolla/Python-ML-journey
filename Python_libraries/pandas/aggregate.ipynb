```python
## aggregate function in pandas
## aggregate() method allows you to apply a function or a list of function names to bo executed along onew of the axis of dataframe,default is 0 (i.e row)
```

```python
## agg() is alias for aggregate()
```

```python
## syntax
## dataframe.aggreagte(func,axis,args,kwargs)
## func it requries func name or list of function names to aplly to dataframe
## axis 0-row  1-columns
##  return value is series or datafarme object with changes
## most frequently used aggragations are sum,min,max 
```

```python
import pandas as pd
import numpy as np
```

```python
data={
    "x":[50,40,30],
    "y":[300,1112,42],
    "z":[12,23,34]
}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.aggregate(["sum"])
```

```python
df=pd.read_csv("nba.csv")
df
```

```python
df.groupby("Position").sum()
```

```python
df=df.dropna()
```

```python
df
```

```python
df.agg("min")
```

```python
df.agg(["min","max","sum"],axis=0)
```
