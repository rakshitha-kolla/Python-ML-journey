## 1.Droping missing data using dropna()

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Titanic-Dataset.csv")
data
```

```python
data.shape
```

```python
# drop rows containing null values
data.dropna()
```

```python
# drop desired column
data.drop("Pclass",axis=1)
```

```python
# drop desired row
data.drop(1,axis=0)
```

```python
# drop columns having null values
data.dropna(axis="columns")
```

```python
## counts number of non null rows
data.count()
```

```python
# how=any means drops rows contaning any column null 
# how=all drops the rows containing all columns null
data.dropna(how="all")
```

```python
data.dropna()
```

```python
data.dropna(subset=["Cabin"])
```

```python
data.dropna(subset=["Embarked"])
```

```python
data.dropna(subset=["Cabin","Embarked"])
```

## 2.Filling Missing Values using fillna()

```python
## fills NaN with value
```

```python
data.fillna(0)
```

```python
## fill specific column
data["Cabin"].fillna("Not Specified", inplace = True)
```

```python
data
```

```python
data["Age"].fillna(30)
```

```python
data["Age"]
```

```python
data["Age"].fillna(method="ffill")
```

```python
data["Age"].fillna(method="bfill")
```
