## Find the most destructive over batsman in the history
## death over last 5 overs
## strike rate=(number of runs/number of balls)/100
## min batsman 200 balls in over 16-20

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("deliveries.csv")
```

```python
df
```

```python
death_over=df[df["over"]>15]
```

```python
death_over
```

```python
all_batsman=death_over.groupby("batsman")['batsman_runs'].count()
x=all_batsman>200
list=all_batsman[x].index.tolist()
```

```python
list
```

## if there is single value we can use mask=df[]==value
## if there is a list we can use df[].isin(list)

```python

mask=death_over["batsman"].isin(list)
```

```python
death_over=death_over[mask]
```

```python
death_over
```

## To calc strike rate we need to find
## runs scored by batsman
## balls played by batsman

```python
runs=death_over.groupby("batsman")["batsman_runs"].sum()
runs
```

```python
balls=death_over.groupby("batsman")["batsman_runs"].count()
balls
```

```python
strike=(runs/balls)*100
```
