## set_index() and reset_index()

```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv("matches.csv")
match.head()
```

```python
## here using id we can uniqly identify each row but in pandas by default index is generated to set id as index we use set_index() function
## we pass key as "id" so id becomes idex and in column field id is not shown
```

```python
print(match.shape)
print(match.set_index("id").shape)
```

```python
match.set_index("id",inplace=True)
```

```python
match
```

```python
match.reset_index(inplace=True)
```
